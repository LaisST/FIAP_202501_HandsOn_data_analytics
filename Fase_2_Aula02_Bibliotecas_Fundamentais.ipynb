## Como começar a construir um modelo de ML?
Separar os dados em uma base de dados de teste
Separar os dados em uma base de dados de treino

É possível usar a biblioteca para fazer essa separação.

### Por que é importante a separação?
Treino (train set): usado para "ensinar" o modelo a reconhecer padrões.

Teste (test set): usado para avaliar se o modelo aprendeu de verdade ou apenas "decorou" os dados.

### Como separar?
Na prática, dividimos os dados em percentuais:
Treino: geralmente 70% a 80% dos dados.

Teste: geralmente 20% a 30% dos dados.

A biblioteca mais usada em Python é o scikit-learn (sklearn), que tem a função train_test_split.

```python
# Importar uma parte da biblioteca sklearn para separar a base de dados
from sklearn.model_selection import train_test_split
```

```python
# Separar os dados em variaveis x e y
# O x será a variável das caracteristicas.
# Obs: O parâmetro axis=1 diz que queremos remover uma coluna (se fosse axis=0, removeria uma linha).
x = df_diabetes.drop(['Class variable'], axis=1)
x.head()
```

```python
# Variavel y é a Target (O alvo que queremos atingir)
y = df_diabetes['Class variable']
y.head(5)
```

```python
# Separar a base em base treino e base teste
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
# Mostrar o tamanho das bases
print(f'O Tamanho da base de treino é {len(x_train)} e o tamanho da base de teste é {len(x_test)}')
print(f'O Tamanho da base original é {df_diabetes.shape[0]} linhas e {df_diabetes.shape[1]} Colunas')
```

```python
# Importar a biblioteca para construir o modelo de ML
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Armazenar a função em uma variavel
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
# Treinar o modelo
knn.fit(x_train, y_train)
```

```python
# Testar o modelo
accuracy = knn.score(x_test, y_test)
```

```python
# O retorno de 70% de acerto, consequentemente foram 30% de erros.
# Se esse % está aceitável vai depender do tipo de base e quantidade de dados analisados/treinado
accuracy = round(accuracy* 100, 2)
print(f"O teste do modelo deu {accuracy}% de acertos")
```
